# Download Kaggle Dataset

This notebook loads Kaggle credentials from the project and authenticates the Kaggle API.

```python
import json
import os
from pathlib import Path


def find_kaggle_config(start: Path) -> Path:
    """Search upward for a project-local Kaggle config file."""
    candidate_dirs = ("artfacts", "artifacts")
    for base in (start, *start.parents):
        for directory in candidate_dirs:
            candidate = base / directory / "kaggle.json"
            if candidate.exists():
                return candidate
    raise FileNotFoundError(
        "Could not find a Kaggle config file. Expected one of: "
        "artfacts/kaggle.json or artifacts/kaggle.json, or set "
        "KAGGLE_USERNAME and KAGGLE_KEY in the environment."
    )


kaggle_config_path = find_kaggle_config(Path.cwd())
with kaggle_config_path.open("r", encoding="utf-8") as fh:
    kaggle_config = json.load(fh)

os.environ["KAGGLE_USERNAME"] = kaggle_config["username"]
os.environ["KAGGLE_KEY"] = kaggle_config["key"]

project_root = kaggle_config_path.parent.parent
print(f"Using Kaggle credentials from: {kaggle_config_path}")
```

```python
from kaggle.api.kaggle_api_extended import KaggleApi

api = KaggleApi()
api.authenticate()
print("Kaggle authentication succeeded.")
```

```python
# Fill in the dataset slug before running this cell, for example:
# dataset = "promptcloud/twitter-airline-sentiment"
dataset = ""
download_dir = project_root / "artfacts" / "downloads"
download_dir.mkdir(parents=True, exist_ok=True)

if dataset:
    api.dataset_download_files(dataset, path=download_dir, unzip=True)
    print(f"Downloaded {dataset} into {download_dir}")
else:
    print("Set the `dataset` variable to a Kaggle dataset slug, then rerun this cell.")
```
